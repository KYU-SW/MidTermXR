## 6. Dictionary 를 이용하여 문제 풀것.

한 도서관에서는 학생들의 열람실 좌석 상태를 기록하고 있다.  
각 학생은 현재 다음 3가지 상태 중 하나를 가질 수 있다.

- `study` : 현재 열람실에서 공부 중
- `break` : 잠시 자리를 비움
- `exit` : 열람실을 완전히 나감

현재 `study`, `break`, `exit` 상태인 학생들을 구하고, 각 상태별 인원 수도 함께 출력하는 프로그램을 작성하시오.

### **입력**
첫째 줄에 로그의 개수 `n`이 주어진다. 
다음 `n`개의 줄에는 학생의 이름과 상태가 공백으로 구분되어 주어진다.

상태는 `study`, `break`, `exit` 중 하나이다.

학생의 이름은 알파벳 대소문자로 이루어진 10글자 이하의 문자열이며, 동명이인은 없다

그 다음 줄에 `break` 상태인 학생 수를 출력한다.

이후 `exit` 상태인 학생의 이름을 사전 순으로 한 줄에 한 명씩 출력한다.
